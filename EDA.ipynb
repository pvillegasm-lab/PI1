## Análisis descriptivo

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
import numpy as np

# Carpeta de entrada
CARPETA_ENTRADA = './Processed Data2/'
os.makedirs(CARPETA_ENTRADA, exist_ok=True)

def analisis_descriptivo(carpeta_entrada):
    archivos = [f for f in os.listdir(carpeta_entrada) if f.endswith('.csv')]

    for archivo in archivos:
        ruta_entrada = os.path.join(carpeta_entrada, archivo)
        df = pd.read_csv(ruta_entrada)
        df['timestamp'] = pd.to_datetime(df['timestamp'], errors='coerce')

        print(f"\n📊 Análisis de: {archivo}")
        
        # 1. Estadísticas descriptivas básicas
        print("\n📌 Estadísticas descriptivas:")
        print(df.describe())

        # 2. Conteo de valores faltantes
        print("\n🔎 Valores nulos por columna:")
        print(df.isnull().sum())

        # 3. Histograma de variables numéricas
        df.hist(figsize=(12, 10), bins=30)
        plt.suptitle(f'Distribución de variables numéricas - {archivo}')
        plt.show()
        plt.figure(figsize=(6, 4))
        sns.histplot(df['return_1d'], bins=100, kde=True, color='steelblue')
        plt.title(f'Distribución de retornos diarios - {archivo}')
        plt.xlabel('Retorno diario')
        plt.ylabel('Frecuencia')
        plt.grid(True)
        plt.show()

        # 4. Correlación de Pearson
        columnas_numericas = df.select_dtypes(include=[np.number]).columns.tolist()
        df_numerico = df[columnas_numericas]
        plt.figure(figsize=(12, 8))
        sns.heatmap(df_numerico.corr(), annot=True, fmt='.2f', cmap='coolwarm', vmin=-1, vmax=1, center=0, square=True)
        plt.title(f'Mapa de calor de correlación (Pearson) - {archivo}')
        plt.show()

        # 5. Correlación de Spearman
        plt.figure(figsize=(12, 8))
        sns.heatmap(df_numerico.corr(method='spearman'), annot=True, fmt='.2f', cmap='viridis', vmin=-1, vmax=1, center=0, square=True)
        plt.title(f'Mapa de calor de correlación (Spearman) - {archivo}')
        plt.show()

        # 6. Correlaciones fuertes (>|0.7|)
        corr_matrix = df_numerico.corr().abs()
        high_corr = (corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
                     .stack().reset_index())
        high_corr.columns = ['Variable 1', 'Variable 2', 'Correlación']
        print("\n🔗 Correlaciones mayores a 0.7:")
        print(high_corr[high_corr['Correlación'] > 0.7])

        # 7. Gráfico de dispersión entre sma20 y close
        if 'sma20' in df.columns and 'close' in df.columns:
            sns.scatterplot(data=df, x='sma20', y='close')
            plt.title(f'Relación entre sma20 y precio de cierre - {archivo}')
            plt.show()
        else:
            print("⚠️ No se encontraron las columnas 'sma20' y/o 'close'.")

# Ejecutar el análisis
analisis_descriptivo(CARPETA_ENTRADA)
```
